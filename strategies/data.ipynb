```python
# Order config
STRATEGY_NAME = "nifty_option_sell"
EXCHANGE = "NFO"      # NSE Futures & Options
PRODUCT = "NRML"      # Normal position (carry forward)
PRICE_TYPE = "MARKET" # Market order
QUANTITY = 75         # NIFTY lot size

# Option symbols to sell
option_symbols = "NIFTY14AUG2524800CE"

# Place SELL orders

order = client.placeorder(
    strategy=STRATEGY_NAME,
    symbol=option_symbols,
    action="SELL",
    exchange=EXCHANGE,
    price_type=PRICE_TYPE,
    product=PRODUCT,
    quantity=QUANTITY
)
print(f"Sell order placed for {option_symbols}: {order}")
```

```python
"""
OpenAlgo WebSocket Feed Example
"""

from openalgo import api
import time

# Initialize feed client with explicit parameters
client = api(
    api_key=Api_key,  # Replace with your API key
    host="http://127.0.0.1:5000",  # Replace with your API host
    ws_url="ws://127.0.0.1:8765" 
    # ws_url="wss://expert-capybara-p6x7qwr7gwqf6qqg-5000.app.github.dev//ws"
)

# MCX instruments for testing
instruments_list = [
    {"exchange": "NFO", "symbol": "NIFTY21AUG2524750PE"}
]

def on_data_received(data):
    print("LTP Update:")
    print(data)

# Connect and subscribe
client.connect()
client.subscribe_ltp(instruments_list, on_data_received=on_data_received)

# Poll LTP data a few times
for i in range(100):
    print(f"\nPoll {i+1}:")
    print(client.get_ltp())
    time.sleep(0.5)

# Cleanup
client.unsubscribe_ltp(instruments_list)
client.disconnect()
```
